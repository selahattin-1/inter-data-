```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sıralanmış son 20 veriyi alın
sehir = df["city"].value_counts().sort_values().tail(12)

# Count plot'u oluşturun
plt.figure(figsize=(14, 6))
sns.countplot(x="city", data=df[df["city"].isin(sehir.index)])
plt.title("Top 20 Cities by Count")
plt.xlabel("Count")
plt.ylabel("City")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sıralanmış son 20 veriyi alın
sehir = df["city"].value_counts().sort_values().tail(12)

# Count plot'u oluşturun
plt.figure(figsize=(14, 6))
sns.barplot(y="city", data=df[df["city"].isin(sehir.index)])
plt.title("Top 20 Cities by Count")
plt.xlabel("Count")
plt.ylabel("City")
plt.show()
```
