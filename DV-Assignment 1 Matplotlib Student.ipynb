## Kütüphaneleri import edin

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

## Değişkenlerimizi tanımlayalım

m değerini output a göre numpy kütüphanesi yardımı ile oluşturalım

```python
m = np.arange(0.0, 11.0)
```

```python
m
```

```python
c = 3* 10**8
```

```python
E = m* c**2
```

```python
E
```

## Oluşturduğumuz dokümanları plotlayalım

#### PLOTTING E=mc^2

```python
plt.plot(m , E)

plt.title("e=mc^2")

plt.show()
```
